Make 5 questions and solve them by using group by.


```python
#Q: How many results we have ? 
covid_data.groupby('Study Results').size().reset_index()
```

```python
#Q: What is most studied type ?
covid_data.groupby('Study Type').size().reset_index()
# Interventional is the most studied type 
```

```python
# Q: Who funded the most ? 
covid_data.groupby('Funded Bys').size().reset_index().max()
# After U.S. Fed|Oher , its Industry who funded the most 
```

```python
covid_data.groupby('Conditions')['Study Results'].sum().reset_index()
```

```python
# Distribute phases and show their count
covid_data.groupby('Phases').size().reset_index()
```

```python
# Show distribution of m/f  or All 
covid_data.groupby('Gender').size().reset_index()
```
